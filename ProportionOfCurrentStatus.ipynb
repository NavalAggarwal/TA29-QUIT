```python
import pandas as pd
import warnings
import sklearn

warnings.filterwarnings('ignore')
```

```python
%%time
# Reading the csv files for initial cleaning of the data
df_nsw = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/nsw.xlsx", header=None)
df_vic = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/vic.xlsx", header=None)
df_act = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/act.xlsx", header=None)
df_qld = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/qld.xlsx", header=None)
df_wa = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/wa.xlsx", header=None)
df_nt = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/nt.xlsx", header=None)
df_tas = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/tas.xlsx", header=None)
df_sa = pd.read_excel("datasets/Proportion of current Smoking status/InitialDataset/sa.xlsx", header=None)
```

```python
df_nsw.info()
```

```python
# Method to remove unwanted rows
def remove_unwanted_row(dataframe):
    dataframe = dataframe[1:-2]
    
    # Setting the 1st row as its headers
    dataframe.columns = dataframe.iloc[0]
    dataframe = dataframe[1:]
    
    return dataframe

df_nsw = remove_unwanted_row(df_nsw)
df_vic = remove_unwanted_row(df_vic)
df_act = remove_unwanted_row(df_act)
df_qld = remove_unwanted_row(df_qld)
df_wa = remove_unwanted_row(df_wa)
df_nt = remove_unwanted_row(df_nt)
df_tas = remove_unwanted_row(df_tas)
df_sa = remove_unwanted_row(df_sa)
```

```python
# Data Cleaning
def update_year(dataframe):
    dataframe.loc[dataframe.Year == '2001', 'Year']  = '2001-02'
    
    return dataframe

df_nsw = update_year(df_nsw)
df_vic = update_year(df_vic)
df_act = update_year(df_act)
df_qld = update_year(df_qld)
df_wa = update_year(df_wa)
df_nt = update_year(df_nt)
df_tas = update_year(df_tas)
df_sa = update_year(df_sa)
```
